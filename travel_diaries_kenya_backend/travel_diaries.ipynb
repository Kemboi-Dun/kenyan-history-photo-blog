```python

import requests
from bs4 import BeautifulSoup
from cssutils import parseStyle
import random
```

```python

header = {"User-Agent": 'Mozilla/5.0 '}
# test website url
url = "https://traveldiarieskenya.wordpress.com/2014/10/03/a-look-at-kenya-through-the-years-pictureblog-1914-1990s/comment-page-1/"
```

```python
response = requests.get(url, headers=header)
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
# function to extract header (header content)
def extract_header(soup):
    header_image = soup.find('header', class_="site-header").find('a').find('img')['src']
    title = soup.find('div', class_="site-branding").find('h1', class_='site-title').find('a').text.strip()
    description = soup.find('div', class_='site-branding').find('h2', class_='site-description').text.strip()
    
    return header_image, title, description
    
```

```python
# function to extract images,title, description and copyright
def extract_images(soup):
    images = []
    for image_tag in soup.find_all("img"):
        image_src = image_tag['src']
        image_description = soup.find('div', class_='jp-carousel-photo-info').find('h3', class_='jp-carousel-caption')
        image_title = soup.find('div', class_='jp-carousel-photo-title-container').find('h2', class_='jp-carousel-photo-title')
        image_copyright = soup.find('ul', class_='jp-carousel-image-exif').find('li')
        images.append((image_src,image_description,image_title,image_copyright))
    return images
        
```
